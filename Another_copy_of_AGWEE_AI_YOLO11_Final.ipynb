```python
# Ultralytics ve YOLO11 kurulum kontrolü

import ultralytics
from ultralytics import YOLO
import torch

print("=" * 60)
print("AGWEE.AI — YOLO11 KURULUM KONTROLÜ")
print("=" * 60)

print("Ultralytics sürümü:", ultralytics.__version__)
print("PyTorch sürümü:", torch.__version__)
print("CUDA kullanılabilir mi?:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU modeli:", torch.cuda.get_device_name(0))

ultralytics.checks()
```

```python
# Önceden eğitilmiş YOLO11n Classification modelini yükleme

model = YOLO("yolo11n-cls.pt")

print("=" * 60)
print("YOLO11n Classification modeli başarıyla yüklendi.")
print("=" * 60)

model.info()
```

```python
# Modeli GPU'ya taşıma

device = 0 if torch.cuda.is_available() else "cpu"
model.to("cuda" if torch.cuda.is_available() else "cpu")

print("Kullanılacak cihaz:", "GPU - Tesla T4" if device == 0 else "CPU")
```
